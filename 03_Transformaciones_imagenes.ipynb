# **Introducción**
## Este notebook muestra cómo realizar diversas transformaciones geométricas, de intensidad y espaciales utilizando librerías populares en Python: Pillow, matplotlib y numpy.

## **Librerías utilizadas:**
 - ```Pillow```: Para manipulación de imágenes.
 - ```matplotlib```: Para la visualización de imágenes.
 - ```numpy```: Para operaciones matemáticas y soporte matricial.
 - ```requests```: Para descargar imágenes desde URLs.
 - ```io.BytesIO```: Para manipular datos binarios en memoria como si fueran archivos.


````python
from PIL import Image, ImageEnhance, ImageOps
import matplotlib.pyplot as plt
import numpy as np
import requests
from io import BytesIO
````

# Descargar la imagen desde una URL

````python
# Descargar la imagen desde una URL
url = "https://images.pexels.com/photos/1676914/pexels-photo-1676914.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content))

# Mostrar la imagen original
plt.imshow(image)
plt.axis("off")
plt.title("Imagen Original")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

# **Transformaciones geométricas**


## 1. Image.rotate

Propósito: Ruta la imagen en un ángulo específico.<br>
Uso:

```
rotated_image = image.rotate(angle, expand=True)
```
Parámetros:

- ```angle```: ```(float)``` El ángulo en grados para rotar la imagen.
- ```expand```: ```(bool, opcional)``` Si es ```True```, ajusta el tamaño del lienzo para acomodar la rotación. Por defecto es ```False```.

````python
# Rotación
rotated_image = image.rotate(45, expand=False)  # Rotación sin expansión

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")  # Quitar los ejes para una visualización más limpia
ax[0].set_title("Original") # Título para la imagen

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(rotated_image)
ax[1].axis("off") # Quitar los ejes para una visualización más limpia
ax[1].set_title("Rotada") # Título para la imagen

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

## 2. Image.resize

Propósito: Cambia el tamaño de la imagen.<br>
Uso:

```
scaled_image = image.resize((new_width, new_height))
```
Parámetros:

- ```new_width```: Nuevo ancho de la imagen en píxeles.
- ```new_height```: Nueva altura de la imagen en píxeles.

````python
# Escalado
scaled_image = image.resize((1000, 1000))  # Escalado a un tamaño específico

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(scaled_image)
ax[1].axis("off")
ax[1].set_title("Escalada")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

## 3. Image.transform

Propósito: Realiza transformaciones geométricas avanzadas, como traslaciones o deformaciones.<br>
Uso:

```
transformed_image = image.transform(size, Image.AFFINE, data)
```
Parámetros:

- ```size```: Tamaño de la imagen transformada.
- ```Image.AFFINE```: Define que la transformación será afín.
- ```data```: Coeficientes de la transformación afín en el formato (a, b, c, d, e, f).
  Fórmulas:
  ```
  x' = ax + by + c
  y' = dx + ey + f
  ```

````python
# Traslación
translated_image = image.transform(image.size, Image.AFFINE, (1, 0, 1000, 0, 1, 1000))

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(translated_image)
ax[1].axis("off")
ax[1].set_title("Traslación")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

# **Transformaciones de intensidad**


## 4. ImageEnhance.Brightness

Propósito: Ajusta el brillo de la imagen.<br>
Uso:

```
bright_image = ImageEnhance.Brightness(image).enhance(factor)
```
Parámetros:

- ```factor```: Factor de ajuste del brillo. Valores mayores a 1 aumentan el brillo; valores menores a 1 lo reducen.

````python
# Ajuste de brillo
bright_image = ImageEnhance.Brightness(image).enhance(1.5)

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(bright_image)
ax[1].axis("off")
ax[1].set_title("Brillo")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

## 5. ImageEnhance.Contrast

Propósito: Ajusta el contraste de la imagen.<br>
Uso:

```
contrast_image = ImageEnhance.Contrast(image).enhance(factor)
```
Parámetros:

- ```factor```: Factor de ajuste del contraste. Valores mayores a 1 aumentan el contraste; valores menores a 1 lo reducen.


````python
# Ajuste de contraste
contrast_image = ImageEnhance.Contrast(image).enhance(2.0)  # Duplicar contraste

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(contrast_image)
ax[1].axis("off")
ax[1].set_title("Contraste")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

## 6. ImageOps.invert
Propósito: Invierte los colores de la imagen.<br>
Uso:

```
inverted_image = ImageOps.invert(image.convert("RGB"))
```
Notas:
- La imagen debe estar en modo ```RGB```. Convierte la imagen primero con ```image.convert("RGB")``` si es necesario.



````python
# Inversión de colores
inverted_image = ImageOps.invert(image.convert("RGB"))

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(inverted_image)
ax[1].axis("off")
ax[1].set_title("Inversión")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

# **Transformaciones espaciales**


##  7. Image.crop

Propósito: Recorta un área específica de la imagen.<br>
Uso:

```
cropped_image = image.crop((left, top, right, bottom))
```
Parámetros:

- ```(left, top, right, bottom)```: (tuple) Coordenadas del rectángulo a recortar.


````python
# Dimensiones de la imagen
width, height = image.size
# Recorte
cropped_image = image.crop((1000, 1000, width - 500, height - 1000))  # Recortar una región específica

# Crear un espacio para dos imágenes (1 fila x 2 columnas)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la imagen original en el primer subgráfico
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title("Original")

# Mostrar la imagen transformada en el segundo subgráfico
ax[1].imshow(cropped_image)
ax[1].axis("off")
ax[1].set_title("Recorte")

# En Jupyter Notebooks, no es necesario llamar plt.show()
````

## 8. Image.transform

Propósito: Realiza transformaciones geométricas avanzadas, como traslaciones o deformaciones.<br>
Uso:

```
deformed_image = image.transform(size, Image.AFFINE, data)
```
Parámetros:

- ```size```: Tamaño de la imagen transformada.
- ```Image.AFFINE```: Define que la transformación será afín.
- ```data```: Coeficientes de la transformación afín en el formato (a, b, c, d, e, f).
  Fórmulas:
